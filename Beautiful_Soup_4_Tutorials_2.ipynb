```python
from bs4 import BeautifulSoup
import requests

url = "https://coinmarketcap.com/"
result = requests.get(url).text
doc = BeautifulSoup(result, "html.parser")

tbody = doc.tbody
print(tbody)
```

```python
trs = tbody.contents
print(trs)
```

```python
print(trs[0].next_sibling)
```

```python
print(trs[1].previous_sibling)
```

```python
print(trs[0].next_siblings)
```

```python
print(list(trs[0].next_siblings))
```

```python
print(trs[0].parent)
```

```python
print(trs[0].parent.name)
```

```python
print(list(trs[0].descendants))
```

```python
print(list(trs[0].contents))
```

```python
print(list(trs[0].children))
```

```python
# Look for names of cryptocurrencies
# and prices
```

```python
tbody = doc.tbody
trs = tbody.contents

prices = {}

#tr stands for table row

for tr in trs[:10]:
  name, price = tr.contents[2:4]
  amended_name = name.p.string
  amended_price = price.a.string
  
  prices[amended_name] = amended_price

print(prices)
```
